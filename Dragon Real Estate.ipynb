## Extract Independent and dependent feature

```python
X = housing_df.iloc[:,:-1]
y = housing_df.iloc[:,-1]
```

## Train Test Split dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state = 42)
```

## Creating a pipeline 

```python

# here pipeline created for standardize data we can add multiple operation in this pipeline 

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

myPipeline = Pipeline([
    ('std_scaler',StandardScaler())
])
```

```python
X_train = myPipeline.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = myPipeline.transform(X_test)
```

## Model Training using RegressionModel

```python
from sklearn.linear_model import LinearRegression

regressionModel = LinearRegression()
regressionModel.fit(X_train,y_train)
```

```python
# Print coefficient and intercept

display(regressionModel.coef_)
```

```python
display(regressionModel.intercept_)
```

```python
display(regressionModel.score(X_train,y_train))
```

## checking accuracy based on metrics

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

y_train_predict =  regressionModel.predict(X_train)
display('mae ---> ',mean_absolute_error(y_train_predict,y_train))
display('mse ---->',mean_squared_error(y_train_predict,y_train))
display('rmse --->',np.sqrt(mean_squared_error(y_train_predict,y_train)))
```

## Checking with other different regression model 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

decisionTreeModel = DecisionTreeRegressor()
```

```python
parameters={"criterion":['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
            "splitter":["best","random"],
            "max_depth" : [1,3,5,7,9,11,12],
            "max_features":["auto","log2","sqrt",None]}
```

```python
decisionTuningModel = GridSearchCV(decisionTreeModel ,param_grid = parameters,cv=5,scoring='neg_mean_squared_error')
```

```python
decisionTuningModel.fit(X_train,y_train)
```

```python
decisionTuningModel.best_params_
```

## Tunning decision tree with best parameters

```python
decision_tuned_hyper_model= DecisionTreeRegressor(criterion= 'squared_error',max_depth=9,max_features= None,splitter='random')
```

```python
decision_tuned_hyper_model.fit(X_train,y_train)
```

```python

y_train_predict =  decision_tuned_hyper_model.predict(X_train)
display('mae ---> ',mean_absolute_error(y_train_predict,y_train))
display('mse ---->',mean_squared_error(y_train_predict,y_train))
display('rmse --->',np.sqrt(mean_squared_error(y_train_predict,y_train)))
```

```python
# Plotting actual and predicted values 
plt.scatter(y_train,y_train_predict,color = '#EB984E')
```

```python
decision_tuned_hyper_model.score(X_train,y_train)
```

## Random Forest Regressor 

## Tunning decision tree with best parameters¶

```python
from sklearn.ensemble import RandomForestRegressor


randomForestRegressorModel = RandomForestRegressor()
```

```python
parameters={"n_estimators":[90,95,100,105,110,120,150],
            "criterion":["squared_error", "absolute_error", "poisson"],
            "max_features":["log2","sqrt",None]}
```

```python
randomForestTuningModel = GridSearchCV(randomForestRegressorModel,param_grid = parameters,cv=5,scoring='neg_mean_squared_error')
```

```python
randomForestTuningModel.fit(X_train,y_train)
```

```python
randomForestTuningModel.best_params_
```

## RandomForestRegressor with best parameters

```python
randomForestRegressorModel = RandomForestRegressor(criterion='absolute_error', max_features= 'log2', n_estimators= 110)
```

```python
randomForestRegressorModel.fit(X_train,y_train)
```

```python
randomForestRegressorModel.score(X_train,y_train)
```

## Checking error in model with metrics

```python
y_train_predict =  randomForestRegressorModel.predict(X_train)
display('mae ---> ',mean_absolute_error(y_train_predict,y_train))
display('mse ---->',mean_squared_error(y_train_predict,y_train))
display('rmse --->',np.sqrt(mean_squared_error(y_train_predict,y_train)))
```

```python
plt.scatter(y_train,y_train_predict,color = '#117A65')
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_train,y_train_predict)
```

```python
score
```

## Adjusted R2 = 1 - [(1-R2)*(n-1) / (n-k-1)]

```python
## Adjusted R2 

1 - (1 - score) * (len(y_train) -1) / (len(y_train) - X_train.shape[1] - 1)
```

## Apply model on Test Data

```python
y_test_predicted = randomForestRegressorModel.predict(X_test)
```

```python

display('mae ---> ',mean_absolute_error(y_test_predicted,y_test))
display('mse ---->',mean_squared_error(y_test_predicted,y_test))
display('rmse --->',np.sqrt(mean_squared_error(y_test_predicted,y_test)))
```
